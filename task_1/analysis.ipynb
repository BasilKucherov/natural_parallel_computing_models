# Exp 1: Fixed P, Change particles number

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    "Кол-во процессов": [8] * 7,
    "Кол-во частиц": [10**i for i in range(3, 10)],
    "Линейное": [0.000021 * 10**i for i in range(0,7)],
    "Время": [0.000021, 0.000274, 0.002104, 0.021607, 0.193785, 1.980204, 19.534486],
    "Линейное ускорение": [1, 10, 1e2, 1e3, 1e4, 1e5, 1e6],
    "Эффективность 100%": [1] * 7,
}
```

```python
df = pd.DataFrame(data)
df["Ускорение"] = df["Время"] / df["Время"][0]
df["Эффективность"] = df["Ускорение"] / df["Линейное ускорение"]
df
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(x="Кол-во частиц", y="Время", data=df, label="Эксперимент")
sns.lineplot(x="Кол-во частиц", y="Линейное", data=df, label="Линейное")
plt.yscale('log')
plt.xscale('log')

plt.xticks(df['Кол-во частиц'])
plt.title("Зависимость времени выполнения от количества частиц (8 процессов)")
plt.savefig('exp1_time.png', dpi=300)
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(x="Кол-во частиц", y="Эффективность", data=df, label="Эксперимент")
sns.lineplot(x="Кол-во частиц", y="Эффективность 100%", data=df, label="Линейное")
plt.xscale('log')
plt.xticks(df['Кол-во частиц'])
plt.title("Зависимость эффективности от количества частиц (8 процессов)")
plt.savefig('exp1_efficiency.png', dpi=300)
```

# Exp 3: N = 10^3 * P

```python
data = {
    "Кол-во процессов": [1,2,3,4,5,6,7,8,16],
    "Кол-во частиц": [1000 * i for i in [1,2,3,4,5,6,7,8,16]],
    "Время": [0.000175, 0.00017, 0.000168, 0.000173, 0.000173, 0.000159, 0.000159, 0.000164, 0.000377],
    "Линейное": [0.000175] * 9,
    "Линейное ускорение": [1,2,3,4,5,6,7,8,16],
    "Эффективность 100%": [1] * 9,
}
```

```python
df = pd.DataFrame(data)
df["Ускорение"] = df["Время"][0] / df["Время"] * df["Кол-во частиц"] / df["Кол-во частиц"][0]
df["Эффективность"] = df["Ускорение"] / df["Линейное ускорение"]
df
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(x="Кол-во процессов", y="Время", data=df, label="Эксперимент")
sns.lineplot(x="Кол-во процессов", y="Линейное", data=df, label="Линейное")

plt.xticks(df['Кол-во процессов'])
plt.title("Зависимость времени выполнения от количества процессов (N = 1e3 * P)")
plt.savefig('exp3_time.png', dpi=300)
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(x="Кол-во процессов", y="Ускорение", data=df, label="Эксперимент")
sns.lineplot(x="Кол-во процессов", y="Линейное ускорение", data=df, label="Линейное")
# plt.yscale('log')
# plt.xscale('log')

plt.xticks(df['Кол-во процессов'])
plt.title("Зависимость ускорения от количества процессов (N = 1e3 * P)")
plt.savefig('exp3_speedup.png', dpi=300)
```

```python
plt.figure(figsize=(10, 6))

sns.lineplot(x="Кол-во процессов", y="Эффективность", data=df, label="Эксперимент")
sns.lineplot(x="Кол-во процессов", y="Эффективность 100%", data=df, label="Линейное")

plt.xticks(df['Кол-во процессов'])
plt.title("Зависимость эффективности от количества процессов (N = 1e3 * P)")
plt.savefig('exp3_efficiency.png', dpi=300)
```
