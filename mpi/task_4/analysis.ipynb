```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
# fname = 'build/parallel_1_2000_16.txt'
# fname = 'build/parallel_2_15000_2.txt'
# fname = 'build/parallel_3_2000_8.txt'

fname = 'build/parallel_1_2000_2.txt'
with open(fname, 'r') as f:
    data_2 = json.load(f)

df_2 = pd.DataFrame(data_2)
df_2['index'] = df_2.index * 20 + 20


fname = 'build/parallel_1_2000_4.txt'
with open(fname, 'r') as f:
    data_4 = json.load(f)

df_4 = pd.DataFrame(data_4)
df_4['index'] = df_4.index * 20 + 20


fname = 'build/parallel_1_2000_8.txt'
with open(fname, 'r') as f:
    data_8 = json.load(f)

df_8 = pd.DataFrame(data_8)
df_8['index'] = df_8.index * 20 + 20

fname = 'build/parallel_1_2000_16.txt'
with open(fname, 'r') as f:
    data_16 = json.load(f)

df_16 = pd.DataFrame(data_16)
df_16['index'] = df_16.index * 20 + 20
```

```python
fig, axes = plt.subplots(4, 1, figsize=(16, 20))

df_2.plot(x='index', y=list(df_2.columns).remove('index'), kind='line', ax=axes[0])
axes[0].set_title('2 процесса')
axes[0].set_xlabel("Iteration", fontsize=18)
axes[0].set_ylabel("Fraction", fontsize=18)
axes[0].tick_params(axis='x', labelsize=18)
axes[0].tick_params(axis='y', labelsize=18)

df_4.plot(x='index', y=list(df_4.columns).remove('index'), kind='line', ax=axes[1])
axes[1].set_title('4 процесса')
axes[1].set_xlabel("Iteration", fontsize=18)
axes[1].set_ylabel("Fraction", fontsize=18)
axes[1].tick_params(axis='x', labelsize=18)
axes[1].tick_params(axis='y', labelsize=18)

df_8.plot(x='index', y=list(df_8.columns).remove('index'), kind='line', ax=axes[2])
axes[2].set_title('8 процессов')
axes[2].set_xlabel("Iteration", fontsize=18)
axes[2].set_ylabel("Fraction", fontsize=18)
axes[2].tick_params(axis='x', labelsize=18)
axes[2].tick_params(axis='y', labelsize=18)

df_16.plot(x='index', y=list(df_16.columns).remove('index'), kind='line', ax=axes[3])
axes[3].set_title('16 процессов')
axes[3].set_xlabel("Iteration", fontsize=18)
axes[3].set_ylabel("Fraction", fontsize=18)
axes[3].tick_params(axis='x', labelsize=18)
axes[3].tick_params(axis='y', labelsize=18)

fig.suptitle('Распределение нагрузки по процессам в зависимости от итерации', fontsize=20)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig('1.png')
plt.show()
```
