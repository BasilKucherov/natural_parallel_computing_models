```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
def plot(function_name):
    df_parallel = pd.read_csv(f"build/{function_name}_parallel.csv", delimiter=",")
    df_sequential = pd.read_csv(f"build/{function_name}.csv", delimiter=",")

    plt.figure(figsize=(10, 5))

    plt.plot(df_parallel["generation"], df_parallel["bestError"], label="Parallel best", color="red")
    plt.plot(df_parallel["generation"], df_parallel["meanError"], label="Parallel mean", color="blue")

    plt.plot(df_sequential["generation"], df_sequential["bestError"], label="Sequential best", color="red", alpha=0.5)
    plt.plot(df_sequential["generation"], df_sequential["meanError"], label="Sequential mean", color="blue", alpha=0.5)
    plt.title(f"{function_name} (n=10)")
    plt.xlabel("Generation")
    plt.ylabel("Error (MAE)")
    plt.legend()
    plt.savefig(f"build/{function_name}.png")
    plt.show()
```

```python
plot("spherical")
```

```python
plot("rosenbrock")
```

```python
plot("rastrigin")
```
