```python
import pandas as pd
import matplotlib.pyplot as plt

# baseline - sequential (1 process)
n = [100, 200, 400, 800, 1600]
T = [0.228, 0.838, 3.33, 13.185, 73.539]

# parallel - 1 process
n_p1 = [100, 200, 400, 800, 1600]
T_p1 = [0.362, 1.434, 5.621, 22.362, 89.513]

# parallel - 4 processes
n_p4 = [100, 200, 400, 800, 1600]
T_p4 = [0.105, 0.370, 1.535, 5.645, 22.529]

# parallel - 9 processes
n_p9 = [34 * 3, 67 * 3, 134 * 3, 267 * 3, 534 * 3]
t_p9 = [0.103, 0.363, 1.322, 4.832, 16.986]
T_p9 = [t / n1 * n2 for t, n1, n2 in zip(t_p9, n_p9, n)]

```

```python
plt.plot(n, T, label="")
plt.legend()
plt.title("Последовательная версия")
plt.xlabel("Размерность")
plt.ylabel("Время выполнения (с)")
plt.xticks(n)
plt.grid()
plt.savefig("sequential_time.png")
plt.show()
```

```python
plt.plot(n, T_p1, label="")
plt.legend()
plt.title("Параллельная версия (1 процесс)")
plt.xlabel("Размерность")
plt.ylabel("Время выполнения (с)")
plt.xticks(n)
plt.grid()
plt.savefig("parallel_1_time.png")
plt.show()
```

```python
plt.plot(n, T_p4, label="")
plt.legend()
plt.title("Параллельная версия (4 процесса)")
plt.xlabel("Размерность")
plt.ylabel("Время выполнения (с)")
plt.xticks(n)
plt.grid()
plt.savefig("parallel_4_time.png")
plt.show()
```

```python
plt.plot(n, T_p9, label="")
plt.legend()
plt.title("Параллельная версия (9 процессов)")
plt.xlabel("Размерность")
plt.ylabel("Время выполнения (с)")
plt.xticks(n)
plt.grid()
plt.savefig("parallel_9_time.png")
plt.show()
```

```python
speedup_1 = [T[i] / T_p1[i] for i in range(len(n))]
speedup_seq = [T[i] / T[i] for i in range(len(n))]

plt.plot(n, speedup_seq, label="Последовательная версия")
plt.plot(n, speedup_1, label="Параллельная версия (1 процесс)")
plt.legend()
plt.title("Ускорение (1 процесс)")
plt.xlabel("Размерность")
plt.ylabel("Ускорение")
plt.xticks(n)
plt.grid()
plt.savefig("parallel_1_speedup.png")
plt.show()
```

```python
speedup_4 = [T[i] / T_p4[i] for i in range(len(n))]
speedup_seq = [T[i] / T[i] for i in range(len(n))]

plt.plot(n, speedup_seq, label="Последовательная версия")
plt.plot(n, speedup_4, label="Параллельная версия (4 процесса)")
plt.legend()
plt.title("Ускорение (4 процесса)")
plt.xlabel("Размерность")
plt.ylabel("Ускорение")
plt.xticks(n)
plt.grid()
plt.savefig("parallel_4_speedup.png")
plt.show()
```

```python
speedup_9 = [T[i] / T_p9[i] for i in range(len(n))]
speedup_seq = [T[i] / T[i] for i in range(len(n))]

plt.plot(n, speedup_seq, label="Последовательная версия")
plt.plot(n, speedup_9, label="Параллельная версия (9 процессов)")
plt.legend()
plt.title("Ускорение (9 процессов)")
plt.xlabel("Размерность")
plt.ylabel("Ускорение")
plt.xticks(n)
plt.grid()
plt.savefig("parallel_9_speedup.png")
plt.show()
```

```python
efficiency_1 = [speedup_1[i] / 1 for i in range(len(n))]

plt.plot(n, efficiency_1, label="Параллельная версия (1 процесс)")
plt.legend()
plt.title("Эффективность (1 процесс)")
plt.xlabel("Размерность")
plt.ylabel("Эффективность")
plt.xticks(n)
plt.grid()
plt.savefig("parallel_1_efficiency.png")
plt.show()
```
